```python
df = pd.read_csv("/content/drive/MyDrive/Individual_dataset_modified_3.csv")
```

```python
df
```

```python
X = df.iloc[:,67:70]
X["Nationality(/ies)"] = df.iloc[:,4]
Y = df.iloc[:,70]
X["SOW"] = df.iloc[:,7]
X["Product"] = df.iloc[:,8]
X
```

```python
print(X['Nationality(/ies)'].value_counts())
print(X['SOW'].value_counts())
print(X['Product'].value_counts())
```

```python
X = pd.get_dummies(X,columns = ['Nationality(/ies)','SOW','Product'])
```

```python
X
```

```python
from pandas.core.common import random_state
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size = 0.3, random_state=10,shuffle = True)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score
DT = DecisionTreeClassifier(max_depth = 2)
RND = RandomForestClassifier(n_estimators=1000, max_leaf_nodes=16, n_jobs=-1)
DT.fit(X_train,y_train)
RND.fit(X_train,y_train)
y_pred = DT.predict(X_test)
y_pred1 = RND.predict(X_test)
print(accuracy_score(y_test,y_pred))
print(accuracy_score(y_test,y_pred1))
print(f1_score(y_test,y_pred,average='micro'))
print(f1_score(y_test,y_pred1,average='micro'))
```

```python
from sklearn.metrics import balanced_accuracy_score, roc_auc_score, make_scorer
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
# FOR DT MODEL
disp = ConfusionMatrixDisplay.from_estimator(
        DT,
        X_test,
        y_test,
        values_format='d'  
        )
```

```python
# FOR RNF model
disp = ConfusionMatrixDisplay.from_estimator(
        RND,
        X_test,
        y_test,
        values_format='d'
    )
```
